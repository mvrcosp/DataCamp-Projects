## Password Requirments:

Must be at least 10 characters in length

```python
logins['size_greater_10'] = logins.password.str.len() >= 10
logins.head(25)
```

```python
valid_psswrds = logins[logins['size_greater_10']][['id', 'username', 'password']]
bad_psswrds = logins[~logins['size_greater_10']][['id', 'username', 'password']]
```

```python
print(logins.shape, valid_psswrds.shape, bad_psswrds.shape)
```

## Password Requirments:

Must contain at least:

one lower case letter

one upper case letter

one numeric character

one non-alphanumeric character

```python
#lower case letter
valid_psswrds['lowercase'] = valid_psswrds.password.str.contains(r'[a-z]')
valid_psswrds['uppercase'] = valid_psswrds.password.str.contains(r'[A-Z]')
valid_psswrds['numeric'] = valid_psswrds.password.str.contains(r'[0-9]')
valid_psswrds['nonalpha'] = valid_psswrds.password.str.contains(r'\W')
```

```python
valid_psswrds.head(20)
```

```python
check_req = (valid_psswrds['lowercase'] & valid_psswrds['uppercase'] & valid_psswrds['numeric'] & valid_psswrds['nonalpha'])
bad_psswrds = bad_psswrds.append(valid_psswrds[~check_req]).drop(['lowercase', 'nonalpha', 'numeric', 'uppercase'], axis=1)
vldpsswrds = valid_psswrds[check_req][['id', 'username', 'password']]
```

```python
print(vldpsswrds.shape, bad_psswrds.shape)
```

## Password Requirments:

Must not contain the phrase password (case insensitive)

```python
bannedphrase = valid_psswrds.password.str.contains('password', case=False)
bad_psswrds = bad_psswrds.append(vldpsswrds[bannedphrase])
```

```python
print(len(bad_psswrds))
```

```python
vldpsswrds = vldpsswrds[~bannedphrase]
print(len(vldpsswrds))
```

## Password Requirments:

Must not contain the user's first or last name, e.g., if the user's name is John Smith, then SmItH876! is not a valid password.

```python
vldpsswrds['first_name'] = logins['username'].str.extract(r'(\w+)', expand=True)
vldpsswrds['last_name'] = logins['username'].str.extract(r'(\w+$)', expand=True)
```

```python
vldpsswrds
```

```python
for idx, row in vldpsswrds.iterrows():
    if row.first_name in row.password.lower() or row.last_name in row.password.lower():
        vldpsswrds = vldpsswrds.drop(index=idx)
        bad_psswrds = bad_psswrds.append(row, ignore_index=True)
```

```python
print(vldpsswrds.shape, bad_psswrds.shape)
```

```python
bad_pass = round(len(bad_psswrds)/len(logins), 2)
bad_pass
```
