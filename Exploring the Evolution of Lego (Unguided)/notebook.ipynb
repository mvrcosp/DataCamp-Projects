```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import numpy as np
```

```python
sets = pd.read_csv('datasets/lego_sets.csv')
parent = pd.read_csv('datasets/parent_themes.csv')
```

```python
print(parent.shape)
print(sets.shape)
```

```python
print(sets.isnull().sum())
print('\n')
print(parent.isnull().sum())
```

```python
sets.drop('num_parts', axis=1, inplace=True)
sets.dropna(subset=['set_num'], inplace=True)
```

```python
print(sets.isnull().sum())
print('\n')
print(parent.isnull().sum())
```

```python
parentlicensed = parent[parent['is_licensed']]
parentlicensed
```

```python
sets['licensed'] = sets['parent_theme'].isin(parentlicensed.name)
```

```python
sets
```

```python
sets.licensed.value_counts()
```

```python
setslicensed = sets[sets['licensed']]
```

```python
setslicensed
```

```python
setslicensed.parent_theme.value_counts(normalize=True)
```

```python
starwars = setslicensed[setslicensed['parent_theme'] == 'Star Wars']
len(starwars)
```

```python
the_force = int((len(starwars) / len(setslicensed))*100)
the_force
```

```python
setslicensed.year.value_counts()
```

```python
pivot = setslicensed.pivot_table('set_num', index='year', columns='parent_theme', aggfunc='count', margins=True)
pivot
```

```python
popularity = pivot['Star Wars'] / pivot['All']
popularity
```
